```python
import numpy as np 
print("Matrix Operation")
```

```python
a=np.array([[2,4], 
[6,3]]) 
b=np.array([[1,2], 
[3,5]])
```

```python
c=a+b 
d=a-b 
e=a*b 
f=np.dot(a,b) 
a_transpose=a.T 
```

```python
print("\nMatrix A:\n",a)
```

```python
print("\nMatrix B:\n",b)
```

```python
print("\nAddition:\n",c)
```

```python
print("\nSubstraction:\n",d)
```

```python
print("\nMultiplication (Element-wise):\n",e)
```

```python
print("\nMultiplication (Row x Column):\n",f)
```

```python
print("\nTranspose:\n",a_transpose)
```
